```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = np.random.randn(50)
data
```

```python
plt.plot(data)
plt.show()
```

```python
plt.subplot(2,3,1)
plt.subplot(2,3,2)
plt.subplot(2,3,3)
plt.subplot(2,3,4)
plt.subplot(2,3,5)
plt.subplot(2,3,6)
plt.show()
```

```python
hist_data = np.random.randn(100)
scat_data = np.arange(100)
```

```python
plt.subplot(2,2,1)
plt.plot(data)
plt.subplot(2,2,2)
plt.hist(hist_data, bins=20)
plt.subplot(2,2,3)
plt.scatter(scat_data, hist_data)
plt.show()
```

```python
plt.plot(data, 'g+')
plt.show()
```

```python
plt.plot(data, 'rs')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.plot(data)
plt.subplot(2,2,2)
plt.hist(hist_data, bins=20)
plt.subplot(2,2,3)
plt.scatter(scat_data, hist_data)
plt.show()
```

```python
data = np.random.randn(30)
plt.plot(data, 'k--', label='default')
plt.plot(data, 'k--', drawstyle='steps-post', label='steps-pos')
plt.legend()
plt.show()
```

```python
data = np.random.randn(100)
plt.title('random')
plt.xlabel('index')
plt.ylabel('height')
plt.plot(data, 'rx')
plt.show()
```

```python
data = np.random.randn(100)
plt.title('random')
plt.xlabel('index')
plt.ylabel('height')
plt.plot(np.random.randn(100).cumsum(), 'r^', label='one')
plt.plot(np.random.randn(100).cumsum(), 'g', label='two')
plt.plot(np.random.randn(100).cumsum(), 'b', label='three')
plt.legend()
plt.show()
```
